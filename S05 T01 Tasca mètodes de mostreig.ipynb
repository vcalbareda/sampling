* SMOTE

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
# se crea un nuevo DF con los torneos. Solo se selecciona la columna torneos para realizar la prueba.
# en un análisis normal habría que eliminar o modificar las columnas con texto.
league = raw_df [['torneo']]
```

```python
league
```

```python
league['torneo'].unique()
```

```python
# Se renombra los torneos. 0 para mundial, 1 para el europeo, 2 para el coreano
for x in league['torneo']:
    if x == 'WORLDS':
        league['torneo'] = league['torneo'].replace(['WORLDS'],0) 
```

```python
for x in league['torneo']:
    if x == 'LEC':
        league['torneo'] = league['torneo'].replace(['LEC'],1) 
```

```python
for x in league['torneo']:
    if x == 'LCS':
        league['torneo'] = league['torneo'].replace(['LCS'],2) 
```

```python
# comprobación de repeticiones por torneo
league['torneo'].value_counts()
```

```python
# Se extra las muestrar de la LEC y la LCS, dejando fuera el mundial.
lec_vs_lcs = league[(league['torneo'] == 1) | (league['torneo'] == 2)]
lec_vs_lcs['torneo'].value_counts()
```

```python
# Smote para la clase minoritiaria
sm = SMOTE(sampling_strategy = 'minority')
x, y = sm.fit_resample(lec_vs_lcs, lec_vs_lcs['torneo'], )
```
